```python
from dateutil import parser
def smart_date_parser(date_str):
    """
    Tries to parse dates with different formats.
    Returns NaT if parsing fails.
    """
    try:
        return parser.parse(str(date_str), dayfirst=True, yearfirst=False)
    except Exception:
        return pd.NaT
```

```python
from scipy import stats
import numpy as np

# Filter for one brand
brand_name = "Apple"
brand_prices = Amazon_10_year_data.loc[
    Amazon_10_year_data['brand'] == brand_name, 
    'discounted_price_inr'
].dropna()

# Population (all other products)
all_prices = Amazon_10_year_data['discounted_price_inr'].dropna()

# Compute mean & std of population
pop_mean = all_prices.mean()
pop_std = all_prices.std()

# Z-test formula
z_stat = (brand_prices.mean() - pop_mean) / (pop_std / np.sqrt(len(brand_prices)))

# Two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print(f"Brand: {brand_name}")
print(f"Z-value: {z_stat:.4f}")
print(f"P-value: {p_value:.6f}")
```

```python
brand_name = "Apple"

brand_mean = brand_prices.mean()
brand_std = brand_prices.std()
all_mean = all_prices.mean()
all_std = all_prices.std()

print(f"Brand mean: {brand_mean}")
print(f"Brand std: {brand_std}")
print(f"All mean: {all_mean}")
print(f"All std: {all_std}")
from scipy import stats

t_stat, p_val = stats.ttest_1samp(brand_prices, all_prices.mean())
print(f"T-value: {t_stat:.4f}, P-value: {p_val:.6f}")
```

```python
import pandas as pd
import numpy as np
import re

# Convert to string
Amazon_10_year_data['customer_rating'] = Amazon_10_year_data['customer_rating'].astype(str)

# Extract the first number (handles '4 stars', '3/5', '4.5/5.0', etc.)
Amazon_10_year_data['customer_rating'] = (
    Amazon_10_year_data['customer_rating']
    .str.extract(r'(\d+\.?\d*)')[0]   # extract number before '/'
    .astype(float)
)

# Normalize fractional ratings (like 3/5 → 3.0)
Amazon_10_year_data.loc[Amazon_10_year_data['customer_rating'] > 5, 'customer_rating'] = np.nan

# Fill missing values with median
Amazon_10_year_data['customer_rating'].fillna(Amazon_10_year_data['customer_rating'].median(), inplace=True)

print("✅ Ratings cleaned and standardized to 1.0–5.0 scale")
Amazon_10_year_data['customer_rating'].head(10)

Amazon_10_year_data['customer_rating'] = (
    Amazon_10_year_data['customer_rating']
    .astype(str)
    .str.extract(r'(\d+\.?\d*)')[0]   # get number part
    .astype(float)
)

# Normalize and handle invalid values
Amazon_10_year_data.loc[Amazon_10_year_data['customer_rating'] > 5, 'customer_rating'] = np.nan

# Fill missing with "No Rating" (string)
Amazon_10_year_data['customer_rating'] = Amazon_10_year_data['customer_rating'].fillna("No Rating")

print("✅ Ratings standardized (1.0–5.0) and missing filled with 'No Rating'")
Amazon_10_year_data['customer_rating'].head(10)
```

```python
import pandas as pd
import numpy as np
import re

def clean_rating(value):
    """
    Standardize customer ratings to numeric (3, 3.5, 4, 4.5, 5)
    Handles: '3', '3.0', '3/5', '3.0/5.0', '3 stars', etc.
    Keeps NaN as NaN.
    """
    if pd.isna(value):
        return np.nan

    # convert to lowercase string for uniform processing
    value = str(value).strip().lower()

    # extract the first valid number from string
    match = re.search(r'\d+(\.\d+)?', value)
    if match:
        num = float(match.group())
        # convert float like 3.0 → 3, keep 3.5 as is
        return int(num) if num.is_integer() else num
    else:
        return np.nan

Amazon_10_year_data["customer_rating"] = Amazon_10_year_data["customer_rating"].apply(clean_rating)
```
